## Calculation of the Average Age, Sleep Hours Per Day, across the entire dataset: Mean

```python
data = pd.read_csv('heart_attack_prediction_dataset.csv')

Average_Age= sum(data['Age'])/len(data) 
print("Average Age:", Average_Age)
SPH_Average = sum(data['Sleep Hours Per Day'])/len(data) 
print("Average sleep hours per day:", SPH_Average)
BMI_Average = sum(data['BMI'])/len(data) 
print("Average BMI:", BMI_Average)
Heart_Attack_Risk = sum(data['Heart Attack Risk'])/len(data) 
print("Heart Attack Risk:", Heart_Attack_Risk)
```

## Calculation of the median

```python
Sort_Age = sorted(data['Age']) 
'''Sorting the data, so that I can find the median.'''
n = len(Sort_Age)          
if n % 2 == 0:
    median = (Sort_Age[n//2 - 1] + Sort_Age[n/2]) /2 
else:
    median = Sort_Age[n//2]   
print("Median:", median)
```

## Calculating the mode:

```python
mode = data['Age'].mode()[0]
print("Mode:", mode)
```

## Calculation of the weighted mean: 

```python
weight_mean = sum(data['Cholesterol'] * data['Age'])/sum(data['Cholesterol'])
print("Weighted Mean:", weight_mean)
```

## Calculation of the geometric mean:

```python
geo_mean = np.prod(data['Age']) ** (1 / len(data))
print("Geometric Mean:", geo_mean)
```

## Calculation of the Percentiles:

```python
Perc_1st = np.percentile(data['Age'], 25)
print("1st Percentile:", Perc_1st)

Perc_2nd = np.percentile(data['Age'], 50)
print("2nd Percentile:", Perc_2nd)

Perc_3rd = np.percentile(data['Age'], 75)
print("3rd Percentile:", Perc_3rd )
```

## The range: 

```python
Age_Range = max(data['Age']) - min(data['Age'])
print("AgeRange:", Age_Range)
```

## The Interquartile Range: 

```python
Int_Quar_Range = Perc_3rd - Perc_1st
print("InterQuartile Range:", Int_Quar_Range)
```

## The Variance: 

```python
Vari = sum(( v- Average_Age) ** 2 for v in data['Age']) / len(data)
print("Variance:", Vari)
```

## Standard Deviation: 

```python
Standard_Dev = Vari ** 0.5 
print("Standard Deviation:", Standard_Dev)
```

## The Skewness of the data:

```python
skewN = 3 * (Average_Age - median )/ Standard_Dev
print("Skewness:", skewN)
```

## Chebyshev's Theorem: 

```python
c = 2 
prop = 1-1 / c ** 2
print("Chebyshev's Theorem:", prop)
```

## Emperical Rule: 

```python
variable1 = len([z for z in zscore if abs(z) <= 1]) / len(data['Age'])
print("Variability within one standard devation:", variable1) 
variable2 = len([z for z in zscore if abs(z) <= 2]) / len(data['Age'])
print("Variability within two standard devation:", variable2) 
variable3 = len([z for z in zscore if abs(z) <= 3]) / len(data['Age'])
print("Variability within three standard devation:", variable3) 
```

## Outliers: 

```python

outlier = 1.5 * Int_Quar_Range
outs = [r for r in data['Age'] if r < Perc_1st - outlier or r> Perc_3rd + outlier]
print("Outliers:", outs)
```

## Five Number Summary:     

```python
Minimum_Value = min(data['Age'])
print("Minimum Value:", Minimum_Value)
print("Quartile 1",Perc_1st)
print("Median:", median)
print("Quartile 3:",Perc_3rd)
Maximum_Value = max(data['Age'])
print("Maximum Value:", Maximum_Value)
```

## Covariance Calculation: 

```python

Cov1 = np.cov(data['Age'], data['Heart Rate'])[0][1]
Cov2 = np.cov(data['Age'], data['Cholesterol'])[0][1]
Cov3 = np.cov(data['Age'], data['Sleep Hours Per Day'])[0][1]
print("Covariance Calculation between the Age and Heart Rate:", Cov1)
print("Covariance Calculation between the Age and Cholesterol:", Cov2)
print("Covariance Calculation between the Age and Hours of sleep:", Cov3)
```

## Correlation Coefficient: 

```python
Cor1 = np.corrcoef(data['Age'], data['Heart Rate'])[0][1]
Cor2 = np.corrcoef(data['Age'], data['Cholesterol'])[0][1]
Cor3 = np.corrcoef(data['Age'], data['Sleep Hours Per Day'])[0][1]
print("Correlation Coefficient Calculation between the Age and Heart Rate:", Cor1)
print("Correlation Coefficient Calculation between the Age and Cholesterol:", Cor2)
print("Correlation Coefficient Calculation between the Age and Hours of sleep:", Cor3)
```

## Median Absolute Deviation:

```python
mad = np.median(np.abs(data['Age'] - median))
print("Median Absolute Deviation:", mad)
```

## Cross Tabulation: Physical ACtivity and Age

```python
cr_tb = pd.crosstab(data['Age'], data['Physical Activity Days Per Week'])
print("Cross Tabulation:", cr_tb)
```

## The Age Group That Excercises the most: 

```python
tot_phy_act_per_age = data.groupby('Age')['Physical Activity Days Per Week'].sum()
most_active_per_age = tot_phy_act_per_age.idxmax()
print("The Age group with the highest physical activity:", most_active_per_age)

average_age_act = data.groupby('Age')['Physical Activity Days Per Week'].mean()
avg_age_act = average_age_act.loc[most_active_per_age]
print("The average times they exercise per week:", avg_age_act)
```
